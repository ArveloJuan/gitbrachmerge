Here are the basic paremeters in the way they are presented in the paper in their pairs shown in the x and y tuple of tuples. Sophia Vera

```python
r = (0, -60, -30, 8, -10)
i = (0, -12, 14, -50, -18)

x = ((0, 0), (r[1], r[2]), (0, r[3]), (0, r[4]), (0, 0), (0, 0))
y = ((0, 0), (0, i[3]),    (0, i[4]), (i[1], 0), (0, 0), (i[2], 0))

t = np.linspace(0, 2 * np.pi)
```

This evaluates the fourier series as well as a polar flower for a little bit of flair for the elephant. Sophia Vera

```python
# returns a fourier series from a set of coeficients ((A0, B0), (A1, B1), (A2, B2)...)
def fourier_series (coefficients):
    series = 0
    for i in range(len(coefficients)):
        series += coefficients[i][0]*np.cos(i*t) + coefficients[i][1]*np.sin(i*t)
    return series

#returns a polar flower
def flower():
    return 10*np.sqrt(abs(np.sin(3*t)));
```

This takes the functions defined above and plots them into an elephant using bokeh. Sophia Vera

```python
bokeh.io.output_notebook()

    
p = bokeh.plotting.figure(height = 250, width = 275)

p.line(fourier_series(x), fourier_series(y), color="red", line_width=3)
#draws the flower at the approx coordinate of the elephant's nose
p.line(80 + flower()*np.cos(t), 20 + flower()*np.sin(t), color="pink", line_width=3) 

bokeh.io.show(p)
```
