a) **We have PDF $f(y)$ with a max at $y^*$ with derivatives $f'(y)$ and** $f''(y)$.
\begin{align}
log(f(y)) = log(f(y*)) + \frac{(y-y^*)(f'(y^*))}{f(y^*)} + \frac{(y-y^*)^2}{2}*\frac{(f''(y^*))(f(y^*))-(f'(y))^2}{f(y^*)^2}
\end{align}
Since we know that $y^*$ is a maximum point, so $f'(y^*) = 0$. Thus
\begin{align}
log(f(y)) = log(f(y*)) - \frac{(y-y^*)^2}{2}*\frac{(f''(y^*))}{f(y^*)} 
\end{align}
 **The term $y^*$ refers to the location parameter of the distribution which corresponds to the peak's location , and $-\frac{(y-y^*)^2}{2}*\frac{(f''(y^*))}{f(y^*)}$ refers to the scale parameter higher terms are ignored as the difference $(y-y^*)$ is considered to be small.**   

 **This scale parameter conforms to the third term of the taylor expansion of the \begin{align} log(f(y))  \end{align} where the function in the log is what we are approximating but in order to turn it into the proper standard deviation we must take this value which is \begin{align} \frac{1}{σ^2} \\[1em]  \end{align} and then take the reciprocal square root to get the standard deviation which is what our scaling factor is.**

b) **We cannot approximate a distribution as normal using its mean and variance for distributions with heavy tails because the tail gets so large that we can no longer calculate first moments for them as a result it is impossible to find an approriate mean and variance for a heavy tail distribution as the distribution cannot be integrated which is vital for the method.**

**For a Gamma distribution with $$α = 5$$ and $$β = 2$$ we have that based on equating moments its mean  and standard deviation are** $$
\begin{align}

This block defines the points and a general plotting definition which will put both plots on the same graph with a proper label and color. This will make it easier to plot everything by just calling the definition. This code block does all three functions at one time. The ft and trigger help select which of the taylor expansions is of interest and compute the proper numbers for the location and scale parameter.

```python
# The set of points to be used here
t1 = np.linspace(0, 1)
t2 = np.linspace(0, 10)

# Explicit definitions of both functions here so that they be used more easily
# later in the plotting code
def second_deriv_beta(x):
  y = 9*(((8)*(x**7)*(1-x)**9)-2*(9)*x**8*((1-x)**8)+(8)*(x**9)*(1-x)**7)
  norm_y = y/(sp.special.beta(10,10))
  return norm_y

def beta(x):
  return ((x**9*(1-x)**9)/ (sp.special.beta(10,10)))

def second_deriv_gamma(x):
  y = (((2*x)**5)*(4*(x**2)-16*x+12)*math.exp(-2*x))/(x**3)
  norm_y = y/(sp.special.gamma(5))
  return norm_y

def gamma_f(x):
  k = 1/(sp.special.gamma(5))
  return (((2*x)**5)*k*math.exp(-2*x)/ (x))

# This is a plotting function that takes in 2 functions explicitly two plots and
# uses two switches to determine the last calculation to use.
def plot(linspace, function1, function2, ft,trigger, title, yaxis):
    
    # We define a figure and the two lines that belong here based on the functions
    p = bokeh.plotting.figure(
        height = 400,
        width = 600,
        x_axis_label = "x",
        y_axis_label = yaxis,
    )
    
    z2 = p.line(
        x = linspace,
        y = function2
    )

    z1 = p.line(
        x = linspace,
        y = function1,
        color = "orange"
    )

    # This block of conditionals helps to make sure the definition fits all
    # the curves properly
    if 'b' in ft:
      o = (find_peaks(stats.beta.pdf(t1,10, 10),.0000001))
      h = ((o[0]))
      scale_inter = second_deriv_beta(t1[h])
      scale = (- beta(t1[h])/scale_inter)**(.5)

      if 'c' in trigger:
        plot_fun = stats.norm.cdf(t1,t1[h], scale)

      else:
        plot_fun = stats.norm.pdf(t1,t1[h], scale)

    else:
      o = (find_peaks(stats.gamma.pdf(t2,5, scale = 1/2),.0000001))
      h = ((o[0]))
      scale_inter = second_deriv_gamma(t2[h])
      scale = (-gamma_f(t2[h])/scale_inter)**(.5)

      if 'c' in trigger:
        plot_fun = stats.norm.cdf(t2,t2[h], scale)

      else:
        plot_fun = stats.norm.pdf(t2,t2[h], scale)

    z3 = p.line(
      x = linspace,
      y = plot_fun,
      color = 'red'
    )

    # Plot formatting to add a title and a legend
    p.title = title

    legend = bokeh.models.Legend(items = [
        ('Equating Moments', [z2]),
        ('True Distribution', [z1]),
        ('Taylor Approximation', [z3])
    ], location = (0, 265))

    p.add_layout(legend, 'right')
    return p
```
